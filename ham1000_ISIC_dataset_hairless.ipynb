```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Input, RandomFlip, RandomRotation
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm

# Set up GPU
physical_devices = tf.config.list_physical_devices('GPU')
if len(physical_devices) > 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)

# Load and preprocess data
preprocessed_dir = r'ham10000_data\ISIC-images-preprocessed'
metadata_file = r'ham10000_data\ham10000_metadata.csv'

# Load metadata
df = pd.read_csv(metadata_file)

# Filter out rows with missing images
print("Checking for valid images...")
valid_rows = []
for idx, row in tqdm(df.iterrows(), total=len(df)):
    image_path = os.path.join(preprocessed_dir, f"{row['isic_id']}_preprocessed.jpg")
    if os.path.exists(image_path):
        valid_rows.append(row)

df = pd.DataFrame(valid_rows)
print(f"Found {len(df)} valid images out of {len(df)} total images")

# Print class distribution
print("\nClass distribution:")
print(df['diagnosis_1'].value_counts(normalize=True))

# Create a label mapping
label_mapping = {'Benign': 0, 'Malignant': 1}
df['label'] = df['diagnosis_1'].map(label_mapping)

# Split the data
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)
train_df, val_df = train_df, test_df = train_test_split(train_df, test_size=0.2, stratify=train_df['label'], random_state=42)

# Create a custom dataset
class MelanomaDataset(keras.utils.Sequence):
    def __init__(self, dataframe, preprocessed_dir, batch_size=32, image_size=(256, 256), shuffle=True):
        self.dataframe = dataframe
        self.preprocessed_dir = preprocessed_dir
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.indexes = np.arange(len(self.dataframe))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __len__(self):
        return int(np.ceil(len(self.dataframe) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_indexes = self.indexes[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_df = self.dataframe.iloc[batch_indexes]

        batch_images = []
        batch_labels = []

        for _, row in batch_df.iterrows():
            image_path = os.path.join(self.preprocessed_dir, f"{row['isic_id']}_preprocessed.jpg")
            image = tf.keras.preprocessing.image.load_img(image_path, target_size=self.image_size)
            image = tf.keras.preprocessing.image.img_to_array(image)
            image = image / 255.0  # Normalize
            batch_images.append(image)
            batch_labels.append(row['label'])

        return np.array(batch_images), np.array(batch_labels)

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

# Create datasets
train_dataset = MelanomaDataset(train_df, preprocessed_dir)
val_dataset = MelanomaDataset(val_df, preprocessed_dir)
test_dataset = MelanomaDataset(test_df, preprocessed_dir, shuffle=False)

# Model architecture
Augmentation = Sequential([
    RandomFlip("horizontal_and_vertical"),
    RandomRotation(0.2),
])

model = Sequential([
    Input(shape=(256, 256, 3)),
    Augmentation,
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(units=64, activation="relu"),
    Dense(units=1, activation="sigmoid"),  # Change to single output with sigmoid
])

model.summary()

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',  # Binary crossentropy for binary classification
    metrics=[
        tf.keras.metrics.AUC(name='auc'),  # AUC-ROC metric
        tf.keras.metrics.Precision(name='precision'),  # Precision metric
        tf.keras.metrics.Recall(name='recall'),  # Recall metric
    ]
)

# Early stopping callback
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_auc',  # Use 'val_auc' instead of 'val_accuracy'
    patience=3,
    restore_best_weights=True,
)

# Train the model
history = model.fit(
    train_dataset,
    epochs=25,
    validation_data=val_dataset,
    callbacks=[early_stopping]
)

# Evaluate the model
loss, auc, precision, recall = model.evaluate(test_dataset)
print(f"Test Loss: {loss:.4f}")
print(f"Test AUC: {auc:.4f}")
print(f"Test Precision: {precision:.4f}")
print(f"Test Recall: {recall:.4f}")

# Confusion Matrix
true_labels = []
predicted_labels = []

# Modify the prediction loop to handle batches properly
for batch_idx in range(len(test_dataset)):
    images, labels = test_dataset[batch_idx]
    if len(images) > 0:  # Check if batch is not empty
        true_labels.extend(labels)
        preds = model.predict(images, verbose=0)
        # This line converts probabilities greater than 0.5 to 1 (malignant) and less than or equal to 0.5 to 0 (benign).
        # We can adjust the threshold to increase the number of malignant cases predicted. By lowering the threshold from 0.5, 
        # we make the model more sensitive to predicting positive (malignant) cases, which can reduce false negatives but might increase false positives.
        predicted_labels.extend((preds > 0.5).astype(int).flatten())  # Use thresholding for binary classification. 

# Only create confusion matrix if we have predictions
if len(true_labels) > 0 and len(predicted_labels) > 0:
    cm = confusion_matrix(true_labels, predicted_labels)
    cm_matrix = pd.DataFrame(
        data=cm, 
        columns=['Predicted Benign', 'Predicted Malignant'],
        index=['Actual Benign', 'Actual Malignant']
    )

    plt.figure(figsize=(10, 7))
    sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()
else:
    print("No valid predictions to create confusion matrix")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score, f1_score

# Collect true labels and predicted probabilities
true_labels = []
predicted_probs = []

# Collect true labels and predictions
for batch_idx in range(len(test_dataset)):
    images, labels = test_dataset[batch_idx]
    if len(images) > 0:  # Check if batch is not empty
        true_labels.extend(labels)
        preds = model.predict(images, verbose=0)
        predicted_probs.extend(preds.flatten())

true_labels = np.array(true_labels)
predicted_probs = np.array(predicted_probs)

# Calculate Precision-Recall curve
precision, recall, thresholds_pr = precision_recall_curve(true_labels, predicted_probs)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-6)  # Adding small value to prevent division by zero

# Plot Precision, Recall, and F1-score as a function of the threshold
plt.figure(figsize=(12, 6))
plt.plot(thresholds_pr, precision[:-1], 'b--', label='Precision')
plt.plot(thresholds_pr, recall[:-1], 'g--', label='Recall')
plt.plot(thresholds_pr, f1_scores[:-1], 'r-', label='F1-score')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision, Recall, and F1-score vs. Threshold')
plt.legend()
plt.grid(True)
plt.show()

# Determine the ideal threshold based on F1-score (or choose based on your goal)
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds_pr[optimal_idx]
print(f"Optimal Threshold (based on F1-score): {optimal_threshold:.2f}")

# Plot ROC Curve as well
fpr, tpr, thresholds_roc = roc_curve(true_labels, predicted_probs)
roc_auc = roc_auc_score(true_labels, predicted_probs)

plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Evaluate at the chosen threshold
predicted_labels = (predicted_probs > optimal_threshold).astype(int)
from sklearn.metrics import classification_report

print("Classification Report at Optimal Threshold:")
print(classification_report(true_labels, predicted_labels, target_names=['Benign', 'Malignant']))
```
